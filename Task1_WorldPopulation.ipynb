```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('world_population.csv')

print(df.info())
print(df.head())

# Missing values check
print(df.isnull().sum())
```

```python
print("Shape of Dataset:", df.shape)
print("\nColumn Names:", df.columns.tolist())
```

```python
df.describe()
```

```python
top_10 = df.sort_values(by='2022 Population', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(data=top_10, x='Country/Territory', y='2022 Population', palette='viridis')
plt.title('Top 10 Most Populated Countries in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Select top 5 countries by 2022 Population
top_5_countries = df.sort_values(by='2022 Population', ascending=False).head(5)

# Years to compare
years = ['1970 Population', '1980 Population', '1990 Population',
         '2000 Population', '2010 Population', '2015 Population',
         '2020 Population', '2022 Population']

# Plot line chart
plt.figure(figsize=(12,6))
for i in range(len(top_5_countries)):
    plt.plot(years, top_5_countries.iloc[i][years], marker='o', label=top_5_countries.iloc[i]['Country/Territory'])

plt.title('Population Growth (1970 - 2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by Continent and sum 2022 Population
continent_population = df.groupby('Continent')['2022 Population'].sum().reset_index()

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x='Continent', y='2022 Population', data=continent_population, palette='viridis')

plt.title('Total Population by Continent (2022)')
plt.xlabel('Continent')
plt.ylabel('Total Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Density (per km²)', y='Growth Rate', hue='Continent', palette='tab10', s=100)

plt.title('Growth Rate vs Population Density')
plt.xlabel('Population Density (per km²)')
plt.ylabel('Growth Rate')
plt.legend(title='Continent')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Top 10 Countries by 2022 Population
top10 = df.sort_values(by='2022 Population', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top10, y='Country/Territory', x='2022 Population', palette='rocket')

plt.title('Top 10 Most Populated Countries in 2022')
plt.xlabel('Population')
plt.ylabel('Country')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```
